```python
import pandas as pd
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise import accuracy
```

```python
#   
reader = Reader(rating_scale=(1, 10))  # ,    1  5
data = Dataset.load_from_df(new_ratings[['User-ID', 'ISBN', 'Book-Rating']], reader)

#      
trainset, testset = train_test_split(data, test_size=0.2)
```

```python
#  
model = SVD()
model.fit(trainset)

#  
predictions = model.test(testset)
rmse = accuracy.rmse(predictions)
print(f'RMSE: {rmse}')
```

```python
#    
def get_recommendations(user_id, num_recommendations=5):
    book_ids = ratings_data['ISBN'].unique()
    user_ratings = ratings_data[ratings_data['User-ID'] == user_id]['ISBN']
    
    #     
    predictions = []
    for book_id in book_ids:
        if book_id not in user_ratings.values:
            pred = model.predict(user_id, book_id)
            predictions.append((book_id, pred.est))
    
    #   
    recommendations = sorted(predictions, key=lambda x: x[1], reverse=True)[:num_recommendations]
    
    #    
    recommended_books = books_data[books_data['ISBN'].isin([x[0] for x in recommendations])]
    recommended_books['predicted_rating'] = [x[1] for x in recommendations]
    
    return recommended_books
```

```python
N = 276709
recommended_books = get_recommendations(user_id=N)
recommended_books
```
