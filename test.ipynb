```python
import numpy as np
import matplotlib.pyplot as plt

# Генерация массива с двумя признаками, распределенными нормально
np.random.seed(42)
X = np.random.normal(size=(100, 2))

# Применение сингулярного разложения
U, S, Vt = np.linalg.svd(X)

# Центрирование данных для визуализации
X_centered = X - np.mean(X, axis=0)

# Визуализация
plt.figure(figsize=(8, 8))
plt.scatter(X_centered[:, 0], X_centered[:, 1], alpha=0.5)

# Визуализация сингулярных векторов
for i in range(2):
    vector = Vt.T[:, i] * S[i]
    plt.arrow(0, 0, vector[0], vector[1], color='r', width=0.02, head_width=0.1)

plt.title('SVD разложение: главные компоненты')
plt.xlabel('Признак 1')
plt.ylabel('Признак 2')
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Генерация массива с двумя признаками, распределенными нормально
np.random.seed(42)
X = np.random.normal(size=(100, 2))

# Применение сингулярного разложения
U, S, Vt = np.linalg.svd(X)

# Центрирование данных
X_centered = X - np.mean(X, axis=0)

# Новые координаты в пространстве сингулярных векторов
X_new_coords = np.dot(X_centered, Vt.T)

# Визуализация данных в старой и новой системах координат
plt.figure(figsize=(14, 6))

# Оригинальные данные
plt.subplot(1, 2, 1)
plt.scatter(X_centered[:, 0], X_centered[:, 1], alpha=0.5)
plt.title('Оригинальные данные')
plt.xlabel('Признак 1')
plt.ylabel('Признак 2')
plt.grid(True)
plt.axis('equal')

# Данные в новой системе координат
plt.subplot(1, 2, 2)
plt.scatter(X_new_coords[:, 0], X_new_coords[:, 1], alpha=0.5, color='g')
plt.title('Данные в новой системе координат')
plt.xlabel('Новая компонента 1')
plt.ylabel('Новая компонента 2')
plt.grid(True)
plt.axis('equal')

plt.show()
```
