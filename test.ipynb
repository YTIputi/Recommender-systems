```python
import numpy as np
import matplotlib.pyplot as plt

#     ,  
np.random.seed(42)
X = np.random.normal(size=(100, 2))

#   
U, S, Vt = np.linalg.svd(X)

#    
X_centered = X - np.mean(X, axis=0)

# 
plt.figure(figsize=(8, 8))
plt.scatter(X_centered[:, 0], X_centered[:, 1], alpha=0.5)

#   
for i in range(2):
    vector = Vt.T[:, i] * S[i]
    plt.arrow(0, 0, vector[0], vector[1], color='r', width=0.02, head_width=0.1)

plt.title('SVD :  ')
plt.xlabel(' 1')
plt.ylabel(' 2')
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#     ,  
np.random.seed(42)
X = np.random.normal(size=(100, 2))

#   
U, S, Vt = np.linalg.svd(X)

#  
X_centered = X - np.mean(X, axis=0)

#      
X_new_coords = np.dot(X_centered, Vt.T)

#        
plt.figure(figsize=(14, 6))

#  
plt.subplot(1, 2, 1)
plt.scatter(X_centered[:, 0], X_centered[:, 1], alpha=0.5)
plt.title(' ')
plt.xlabel(' 1')
plt.ylabel(' 2')
plt.grid(True)
plt.axis('equal')

#     
plt.subplot(1, 2, 2)
plt.scatter(X_new_coords[:, 0], X_new_coords[:, 1], alpha=0.5, color='g')
plt.title('    ')
plt.xlabel('  1')
plt.ylabel('  2')
plt.grid(True)
plt.axis('equal')

plt.show()
```
